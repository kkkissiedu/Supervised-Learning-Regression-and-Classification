```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('./deeplearning.mplstyle')
```

Variables

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])

print(f'x_train = {x_train}')
print(f'y_train = {y_train}')
```

Number of Training Examples

```python
print(f'x_train.shape = {x_train.shape}')
print(f'y_train.shape = {y_train.shape}')

### m is the number of training examples
m = x_train.shape[0]
###This assigns the first value(row count) to m

print(f'The number of training examples is {m}.')
###OR###
###m = len(x_train)
```

Training Example

```python
i = 0 #(First value)

###Defining variables for the 'i'th training example
x_i = x_train[i]
y_i = y_train[i]

print(f'(x^({i}),y^({i})) = ({x_i}, {y_i})')
```

Plotting the Training Data

```python
###Plotting the Data Points
plt.scatter(x_train,y_train, marker = 'x', c = 'b')
###Setting Graph Title
plt.title('Housing Prices')
###Setting axes labels
plt.xlabel('Size (1000 sqft)')
plt.ylabel('Cost (1000s of $)')
plt.show()
```

Defining the model's Parameters

```python
w = 200
b = 100

print(f'w = {w}')
print(f'b = {b}')
```

Defining f_wb

```python
###f_wb is the function for the Line of Best Fit for the dataset

def compute_model_output(x, w, b):
    '''
    Args:
      x (ndarray (m,)): Data, m examples....Numpy n-dimensional array with shape (m,) 
      w,b (scalar)    : model parameters....Arguments without dimensions, just a magnitude

    Returns:
      y (ndarray (m,)): target values....Numpy n-dimensional array with shape (m,) 
    '''
    m = x.shape[0]
    f_wb = np.zeros(m)  ###Creates an array with length equal to the value of m (the number of training examples) for the predicted values of y
    for i in range(m):
        f_wb[i] = w * x[i] + b

    return f_wb

    

```

Calling the function

```python
tmp_f_wb = compute_model_output(x_train, w, b,)

###Plotting the resulting Line of Best Fit
plt.plot(x_train, tmp_f_wb, c = 'b', label = 'Our Prediction')

###Plotting the Data Points
plt.scatter(x_train, y_train, marker = 'x', c = 'r')
plt.title('Housing Prices')
plt.xlabel('Size (1000 sqft)')
plt.ylabel('Cost (1000s of $)')

####Adding Legend to graph
plt.legend() 

plt.show()
```
